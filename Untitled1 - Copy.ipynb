# Outliers and skew

```python
plt.figure(figsize=(18,10))
df.boxplot();
plt.xticks(rotation=60, size=10)
```

```python
mask = df.dtypes == np.float
float_cols = df.columns[mask]
```

```python
skew_limit = 0.75
skew_vals = df[float_cols].skew()
```

```python
skew_cols = (abs(skew_vals)
             .sort_values(ascending=False)
             .to_frame()
             .rename(columns={0:'Skew'})
             .query('abs(Skew) > {}'.format(skew_limit)))

skew_cols
```

```python
for col in skew_cols.index.values:
    df[col] = df[col].apply(np.log1p) # for coe_ np.expm1(Y)
```

### Change categorical data to dummies

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
```

```python
data_ohc = df.copy()

ohc = OneHotEncoder()

new_dat = ohc.fit_transform(data_ohc[['New_Zone']])

n_cols = ohc.categories_
col_names = ['_'.join(['New_Zone', str(x)]) for x in n_cols[0]]
new_df = pd.DataFrame(new_dat.toarray(),columns=col_names)
new_data = pd.concat([data_ohc, new_df], axis=1)
```

```python
new_data = new_data.drop('New_Zone', axis=1)
```

```python
from time import strptime
new_data['Month_Year'] = [strptime(i, '%b').tm_mon for i in new_data['Month_Year']]
```

```python
n_df = new_data.drop(['Year', 'Date'],axis=1)
```

```python
from sklearn.model_selection import ShuffleSplit
```

```python
n_df = n_df.drop(['Room Revenue', 'Food & Beverage Revenue', 'Other Revenue'], axis=1)
```

```python
X = n_df.drop('Total Revenues', axis=1).columns
y = n_df['Total Revenues']
```

```python
strat_shuf_split = ShuffleSplit(n_splits=5, 
                                          test_size=0.15, 
                                          random_state=42)

train_idx, test_idx = next(strat_shuf_split.split(n_df[X], y))

# Create the dataframes
X_train = n_df.loc[train_idx, X]
y_train = n_df.loc[train_idx, 'Total Revenues']

X_test  = n_df.loc[test_idx, X]
y_test  = n_df.loc[test_idx, 'Total Revenues']
```

```python
from keras.callbacks import ModelCheckpoint
from keras.models import Sequential
from keras.layers import Dense, Activation, Flatten
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
import warnings 
warnings.filterwarnings('ignore')
warnings.filterwarnings('ignore', category=DeprecationWarning)
```

```python
NN_model = Sequential()

# The Input Layer :
NN_model.add(Dense(128, kernel_initializer='normal',input_dim = X_train.shape[1], activation='relu'))

# The Hidden Layers :
NN_model.add(Dense(256, kernel_initializer='normal',activation='relu'))
NN_model.add(Dense(256, kernel_initializer='normal',activation='relu'))
NN_model.add(Dense(256, kernel_initializer='normal',activation='relu'))

# The Output Layer :
NN_model.add(Dense(1, kernel_initializer='normal',activation='linear'))

# Compile the network :
NN_model.compile(loss='mean_absolute_error', optimizer='adam', metrics=['mean_absolute_error'])
NN_model.summary()
```

```python
checkpoint_name = 'Weights-{epoch:03d}--{val_loss:.5f}.hdf5' 
checkpoint = ModelCheckpoint(checkpoint_name, monitor='val_loss', verbose = 1, save_best_only = True, mode ='auto')
callbacks_list = [checkpoint]
```

```python
NN_model.fit(X_train, y_train, epochs=2000, batch_size=32, validation_split = 0.2, callbacks=callbacks_list)
```

```python
wights_file = 'Weights-1499--0.08092.hdf5' # choose the best checkpoint 
NN_model.load_weights(wights_file) # load it
NN_model.compile(loss='mean_absolute_error', optimizer='adam', metrics=['mean_absolute_error'])
```

```python
predictions = NN_model.predict(X_test)
predictions[:,0]
```

```python
real_y = pd.DataFrame(y_test.reset_index()).drop('index',axis=1)
```

```python
show = pd.DataFrame(predictions, columns=['predictions'])
```

```python
to_comp = pd.merge(real_y , show, left_index=True, right_index=True)
```

```python
np.expm1(to_comp)
```

```python
mean_absolute_error(y_test, predictions[:,0])
```

## Mean Absolute Error is bit higher in test more than train, hanse it's seems like the model have high variance (over fit).
### Action: as i can't increase the sample size, i will try to reduce the layers, and add regularization.

```python
from keras import regularizers
from keras.layers.core import Dropout
```

```python
NN_model_1 = Sequential()

# The Input Layer :
NN_model_1.add(Dense(32, kernel_initializer='normal', kernel_regularizer=regularizers.l2(0.01), input_dim = X_train.shape[1], activation='relu'))

# The Hidden Layers :
NN_model_1.add(Dense(64, kernel_initializer='normal', kernel_regularizer=regularizers.l2(0.01), activation='relu'))
NN_model_1.add(Dense(64, kernel_initializer='normal', kernel_regularizer=regularizers.l2(0.01), activation='relu'))
# The Output Layer :
NN_model_1.add(Dense(1, kernel_initializer='normal', kernel_regularizer=regularizers.l2(0.01), activation='relu'))

# Compile the network :
NN_model_1.compile(loss='mean_absolute_error', optimizer='adam', metrics=['mean_absolute_error'])
NN_model_1.summary()
```

```python
checkpoint_name = 'Weights-{epoch:03d}--{val_loss:.5f}.hdf5' 
checkpoint = ModelCheckpoint(checkpoint_name, monitor='val_loss', verbose = 1, save_best_only = True, mode ='auto')
callbacks_list = [checkpoint]
```

```python
NN_model_1.fit(X_train, y_train, epochs=2000, batch_size=32, validation_split = 0.2, callbacks=callbacks_list)
```

```python
wights_file = 'Weights-1867--0.13298.hdf5' # choose the best checkpoint 
NN_model_1.load_weights(wights_file) # load it
NN_model_1.compile(loss='mean_absolute_error', optimizer='adam', metrics=['mean_absolute_error'])
```

```python
predictions = NN_model_1.predict(X_test)
predictions[:,0]
```
